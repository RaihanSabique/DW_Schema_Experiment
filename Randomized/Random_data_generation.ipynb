```python
from random import randint

def random_with_N_digits(n):
    range_start = 10**(n-1)
    range_end = (10**n)-1
    return randint(range_start, range_end)
```

### Time Dimension 

```python
from random import randrange
import datetime 
import random
import time
from datetime import datetime

def randomDate(start, end):
    frmt = '%d-%m-%Y %H:%M:%S'

    stime = time.mktime(time.strptime(start, frmt))
    etime = time.mktime(time.strptime(end, frmt))

    ptime = stime + random.random() * (etime - stime)
    dt = datetime.fromtimestamp(time.mktime(time.localtime(ptime)))
    return dt

# startDate = datetime.datetime(2013, 9, 20,13,00)
time_key = []
date = []
hour = []
day = []
week = []
month = []
year = []
quarter = []

for x in range(1,10000):
#     print(x)
    if(x<10):
        time_key.append('T0000'+str(x))
    elif(x<100):
        time_key.append('T000'+str(x))
    elif(x<1000):
        time_key.append('T00'+str(x))
    else:
        time_key.append('T0'+str(x))
    d = randomDate("20-01-2014 13:30:00", "23-01-2021 04:50:34")
    date.append(d.strftime("%d-%m-%Y %H:%M"))
    hour.append(d.strftime("%H"))
    day.append(d.strftime("%d"))
    month.append(d.strftime("%m"))
    year.append(d.strftime("%Y"))
    m = int(d.strftime("%m"))
    if(m>=1 and m<=3):
        quarter.append('Q1')
    elif(m>=4 and m<=6):
        quarter.append('Q2')
    elif(m>=7 and m<=9):
        quarter.append('Q3')
    else:
        quarter.append('Q4')
    dd = int(d.strftime("%d"))
    if(dd>=1 and dd<=7):
        week.append('1st Week')
    elif(dd>=8 and dd<=14):
        week.append('2nd Week')
    elif(dd>=15 and dd<=21):
        week.append('3rd Week')
    else:
        week.append('4th Week')
    
# print(len(time_key), len(date), len(quarter))

# time_dim.reset_index(drop=True, inplace=True)
# print(time_dim.index)
# print(month)
# print(quarter)
time_dimension=pd.DataFrame(list(zip(time_key, date, hour, day, week, month, quarter, year)), columns=['time_key', 'date', 'hour', 'day', 'week', 'month', 'quarter', 'year'])
# time_dim['time_key'] = pd.Series(time_key, index=time_dim.index)
# time_dim['date'] = pd.Series(date)
# time_dim['hour'] = pd.Series(hour)
# time_dim['day'] = pd.Series(day)
# time_dim['month'] = pd.Series(month)
# time_dim['quarter'] = pd.Series(quarter)
# time_dim['year'] = pd.Series(year)
print(time_dimension)
```
